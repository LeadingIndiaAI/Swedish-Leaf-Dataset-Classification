```
def fd_hu_moments(image):
    image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    feature = cv2.HuMoments(cv2.moments(image)).flatten()
    return feature
```

```
def fd_haralick(image):
    # convert the image to grayscale
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    # compute the haralick texture feature vector
    haralick = mahotas.features.haralick(gray).mean(axis=0)
    # return the result
    return haralick
```

```
def fd_histogram(image, mask=None):
    # convert the image to HSV color-space
    image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
    # compute the color histogram
    hist  = cv2.calcHist([image], [0, 1, 2], None, [bins, bins, bins], [0, 256, 0, 256, 0, 256])
    # normalize the histogram
    cv2.normalize(hist, hist)
    # return the histogram
    return hist.flatten()
```

```
path ='/content/drive/My Drive/Colab Notebooks/dataset'
```

```
import glob
import cv2
from PIL import Image
imgdata=[]
imglabel=[]
for i in glob.glob(path+'/test/*/'):
    print(i)
    for j in glob.glob(i+'*tif'):
        image = cv2.imread(j)
        image = cv2.resize(image, (224,224))
        fv_hu_moments = fd_hu_moments(image)
        fv_haralick=fd_haralick(image)
        fv_histogram = fd_histogram(image)
        res=np.hstack([fv_hu_moments,fv_haralick,fv_histogram])
        imgdata.append(res)
        imglabel.append(int(i[-2]))
```

```
for i in glob.glob(path+'/train/*/'):
    print(i)
    for j in glob.glob(i+'*jpeg'):
        image = cv2.imread(j)
        image = cv2.resize(image, (224,224))
        fv_hu_moments = fd_hu_moments(image)
        fv_haralick=fd_haralick(image)
        fv_histogram = fd_histogram(image)
        res=np.hstack([fv_hu_moments,fv_haralick,fv_histogram])
        imgdata.append(res)
        imglabel.append(int(i[-2]))
```

```
 for i in glob.glob(path+'/val/*/'):
    print(i)
    for j in glob.glob(i+'*jpeg'):
        image = cv2.imread(j)
        image = cv2.resize(image, (224,224))
        fv_hu_moments = fd_hu_moments(image)
        fv_haralick=fd_haralick(image)
        fv_histogram = fd_histogram(image)
        res=np.hstack([fv_hu_moments,fv_haralick,fv_histogram])
        imgdata.append(res)
        imglabel.append(int(i[-2]))
```

```
from sklearn import model_selection
x_train, x_test, y_train, y_test = model_selection.train_test_split(imgdata, imglabel, random_state=42, test_size=.2)
```
